### Exploratory Data Analysis -EDA

```python
# Adding Nenessaring Packages 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../data/raw/data.csv")
data.head()
```

```python
# Overview of the Data
data.info()
```

```python
# Statistical Summary of the  Data
data.describe()
```

```python
# Distribution of Numerical Features... use Histogram

numerical_cols = ['Amount', 'Value']  # Add more if relevant

for col in ['Amount', 'Value']:
    plt.figure(figsize=(6, 4))
    sns.histplot(data[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlim(0, data[col].quantile(0.95))  # Focus on the 95th percentile
    plt.show()

```

```python
# Distribution of Categorical Features.. usecount plotts
categorical_cols = ['ProductCategory', 'ChannelId', 'CurrencyCode', 'FraudResult']

for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=data, x=col, order=data[col].value_counts().index)
    plt.xticks(rotation=45)
    plt.title(f'Distribution of {col}')
    plt.show()
```

### Correlation Analysis

```python
corr = data[numerical_cols].corr()

plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Missing values ..
missing = data.isnull().sum()
missing[missing > 0].sort_values(ascending=False)
```

```python
# Outlier Detection.. Using BoxPlot
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=data, x=col)
    plt.title(f'Outlier Detection for {col}')
    plt.show()
```

##  Key EDA Insights

1. The `Amount` feature is highly right-skewed, with a small number of large transactions. The same way the `Value` variable behaves similarly to Amount, showing strong right-skew!!
2. The majority of transactions come through the `IOS and Android` channel, indicating platform preference.
3. The `FraudResult` is highly imbalanced, with very few fraudulent cases.
4. Certain customers have unusually high transaction frequencythese may influence RFM clustering.
5. Some important features (e.g., `ProviderId`) have missing values and may need imputation or removal.
